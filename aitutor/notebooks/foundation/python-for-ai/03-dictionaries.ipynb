## Part 2: Accessing Values

Two main ways: square brackets `[]` and `.get()` method.

```python
metrics = {
    'epoch': 10,
    'train_loss': 0.345,
    'val_loss': 0.412
}

# Method 1: Square brackets (raises error if key missing)
loss = metrics['train_loss']
print(f"Train loss: {loss}")

# Method 2: get() - safe, returns None if key missing
learning_rate = metrics.get('learning_rate')  # None
dropout = metrics.get('dropout', 0.5)  # 0.5 (default)

print(f"Learning rate: {learning_rate}")  # None
print(f"Dropout: {dropout}")  # 0.5
```

```python
# AI Example: Safe configuration loading
user_config = {
    'learning_rate': 0.001,
    'batch_size': 64
}

# Get with defaults
lr = user_config.get('learning_rate', 0.01)
batch_size = user_config.get('batch_size', 32)
epochs = user_config.get('num_epochs', 10)  # Use default 10

print(f"Config:")
print(f"  Learning rate: {lr}")
print(f"  Batch size: {batch_size}")
print(f"  Epochs: {epochs} (default)")
```

## Part 3: Modifying Dictionaries

```python
# Start with basic config
config = {
    'learning_rate': 0.01,
    'batch_size': 32
}

print(f"Initial: {config}")

# Update existing key
config['learning_rate'] = 0.001
print(f"After update: {config}")

# Add new key
config['num_epochs'] = 50
print(f"After adding: {config}")

# Delete key
del config['batch_size']
print(f"After delete: {config}")
```

```python
# AI Example: Tracking best model
training_state = {
    'current_epoch': 0,
    'best_val_loss': float('inf'),
    'patience_counter': 0
}

# Simulate 5 epochs
for epoch in range(5):
    training_state['current_epoch'] = epoch + 1
    current_loss = 0.5 / (epoch + 1)  # Decreasing loss
    
    if current_loss < training_state['best_val_loss']:
        training_state['best_val_loss'] = current_loss
        training_state['patience_counter'] = 0
        print(f"Epoch {epoch + 1}: New best! Loss = {current_loss:.4f}")
    else:
        training_state['patience_counter'] += 1
        print(f"Epoch {epoch + 1}: No improvement (patience: {training_state['patience_counter']})")

print(f"\nFinal state: {training_state}")
```

## Part 4: Dictionary Methods

```python
config = {
    'learning_rate': 0.001,
    'batch_size': 32,
    'num_epochs': 10
}

# Get all keys
print(f"Keys: {list(config.keys())}")

# Get all values
print(f"Values: {list(config.values())}")

# Get all key-value pairs
print(f"Items: {list(config.items())}")

# Iterate through dictionary
print("\nIterating:")
for key, value in config.items():
    print(f"  {key}: {value}")
```

```python
# AI Example: Print experiment summary
experiment = {
    'model': 'ResNet50',
    'dataset': 'ImageNet',
    'accuracy': 0.92,
    'training_time': 3600
}

print("Experiment Summary:")
print("=" * 40)
for key, value in experiment.items():
    if isinstance(value, float):
        print(f"{key}: {value:.2%}" if value < 1 else f"{key}: {value:.4f}")
    else:
        print(f"{key}: {value}")
```

## Part 5: Nested Dictionaries

```python
# Complete experiment configuration
experiment = {
    'experiment_id': 'exp_001',
    'model': {
        'architecture': 'ResNet',
        'depth': 50,
        'pretrained': True
    },
    'training': {
        'learning_rate': 0.001,
        'batch_size': 32,
        'num_epochs': 50
    },
    'results': {
        'train_acc': 0.95,
        'val_acc': 0.91,
        'test_acc': 0.89
    }
}

# Access nested values
print(f"Experiment: {experiment['experiment_id']}")
print(f"Model: {experiment['model']['architecture']}{experiment['model']['depth']}")
print(f"Learning rate: {experiment['training']['learning_rate']}")
print(f"Test accuracy: {experiment['results']['test_acc']:.2%}")
```

```python
# AI Example: Hyperparameter search results
search_results = [
    {'lr': 0.1, 'batch_size': 32, 'accuracy': 0.75},
    {'lr': 0.01, 'batch_size': 32, 'accuracy': 0.88},
    {'lr': 0.001, 'batch_size': 32, 'accuracy': 0.91},
    {'lr': 0.01, 'batch_size': 64, 'accuracy': 0.90},
]

# Find best configuration
best_config = max(search_results, key=lambda x: x['accuracy'])

print("Hyperparameter Search Results:")
print("=" * 50)
for config in search_results:
    print(f"LR: {config['lr']:.3f}, Batch: {config['batch_size']:2d}, "
          f"Acc: {config['accuracy']:.2%}")

print("\nBest Configuration:")
print(f"  Learning rate: {best_config['lr']}")
print(f"  Batch size: {best_config['batch_size']}")
print(f"  Accuracy: {best_config['accuracy']:.2%}")
```

## Part 6: Tuples

Tuples are immutable sequences. Perfect for fixed dimensions and coordinates.

```python
# Creating tuples
image_shape = (224, 224, 3)  # Height, width, channels
coordinates = (10, 20)
single_element = (42,)  # Note the comma!

# Accessing elements
height, width, channels = image_shape
print(f"Image: {height}x{width}x{channels}")

# Index access
print(f"First element: {image_shape[0]}")
print(f"Last element: {image_shape[-1]}")
```

```python
# AI Example: Batch shapes
batch_shape = (32, 224, 224, 3)  # (batch, height, width, channels)
output_shape = (32, 1000)  # (batch, num_classes)

print(f"Input shape: {batch_shape}")
print(f"Output shape: {output_shape}")
print(f"Batch size: {batch_shape[0]}")
print(f"Num classes: {output_shape[1]}")
```

```python
# Tuples as dictionary keys
model_cache = {
    (224, 224): "resnet_224.pth",
    (384, 384): "resnet_384.pth",
    (512, 512): "resnet_512.pth"
}

# Look up model by input size
input_size = (224, 224)
model_path = model_cache[input_size]
print(f"Model for {input_size}: {model_path}")
```

## Part 7: Choosing Data Structures

Quick reference guide for AI programming.

```python
# Lists: Ordered, mutable sequences
losses = [0.89, 0.76, 0.65, 0.58]  # âœ… Changes over time

# Dictionaries: Named data
config = {'lr': 0.001, 'batch': 32}  # âœ… Named parameters

# Tuples: Immutable sequences
shape = (224, 224, 3)  # âœ… Fixed dimensions

print(f"Losses (list): {losses}")
print(f"Config (dict): {config}")
print(f"Shape (tuple): {shape}")
```

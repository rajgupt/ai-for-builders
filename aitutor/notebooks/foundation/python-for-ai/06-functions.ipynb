## Part 1: Basic Functions

```python
def calculate_accuracy(predictions, labels):
    '''Calculate accuracy.'''
    correct = sum([p == l for p, l in zip(predictions, labels)])
    return correct / len(labels)

preds = [0, 1, 1, 0, 1]
labels = [0, 1, 0, 0, 1]
acc = calculate_accuracy(preds, labels)
print(f'Accuracy: {acc:.2%}')
```

## Part 2: Default Arguments

```python
def create_config(lr=0.001, batch=32, epochs=10):
    return {'lr': lr, 'batch': batch, 'epochs': epochs}

config = create_config(lr=0.01)
print(config)
```

## Part 3: AI Functions

```python
import numpy as np

def normalize_data(data, method='minmax'):
    if method == 'minmax':
        return (data - data.min()) / (data.max() - data.min())
    elif method == 'zscore':
        return (data - data.mean()) / data.std()

pixels = np.array([0, 50, 100, 150, 200, 255])
norm = normalize_data(pixels)
print(f'Normalized: {norm}')
```

## Part 4: Lambda Functions

```python
relu = lambda x: np.maximum(0, x)
x = np.array([-2, -1, 0, 1, 2])
print(f'ReLU: {relu(x)}')
```
